```python
import os
import json
```

```python
def read_all_files_in_directory(key):
    
    files = os.listdir('D:/political podcasts/transcriptions2')
    for file in files:
        if file.startswith('double_scoring'):
            
            print(f"Reading file '{file}'.")
            file_path = os.path.join('D:/political podcasts/transcriptions2', file)
            # Opening JSON file
            input_file = open (file_path)
            json_array = json.load(input_file)

            json_array.sort(key=lambda x: x[key], reverse=True)
            
            for obj in json_array[:5]:
                print(obj)
            print('----------------------------------------------------------------')
```

```python
read_all_files_in_directory('TOXICITY')
```
